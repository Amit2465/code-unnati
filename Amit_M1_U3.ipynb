# Bar Graph

```python
#Netflix
generics=["Horror","Drama","Comedy","Sci-fi","Thriller","Suspense"]
movies=[156,134,178,230,224,145]
plt.bar(generics,movies, color="y")
plt.legend(["movies"], loc="upper left")
plt.xlabel("generics")
plt.ylabel("No of movies")
plt.title("Netflix Movies")
plt.show()
```

# Graphs

```python
theta1=np.arange(360)
theta1_r=np.deg2rad(theta1)
sintheta=np.sin(theta1_r)
plt.xlabel("Degree")
plt.ylabel("Value")
plt.title("Sin Graph")
plt.plot(theta1,sintheta, color="g")
plt.show()
```

```python
theta1=np.arange(360)
theta1_r=np.deg2rad(theta1)
costheta=np.cos(theta1_r)
plt.xlabel("Degree")
plt.ylabel("Value")
plt.title("Cos Graph")
plt.plot(theta1,costheta, color="g")
plt.show()
```

```python
theta1=np.arange(360)
theta1_r=np.deg2rad(theta1)
sintheta=np.sin(theta1_r)
plt.xlabel("Degree")
plt.ylabel("Value")
plt.title("Sin-cos Graph")
plt.plot(theta1,sintheta, color="g")
theta1=np.arange(360)
theta1_r=np.deg2rad(theta1)
costheta=np.cos(theta1_r)
plt.plot(theta1,costheta, color="y")
plt.legend({"sin","cos"})
plt.annotate("maxima",xy=(90,1))
plt.savefig("myplot.pdf")
plt.show()

```
